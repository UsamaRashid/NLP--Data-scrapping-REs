```python
import nltk
import re
nltk.download('all')
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
  
# link for extract html data
def getdata(url):
    r = requests.get(url)
    return r.text
  
htmldata = getdata("https://en.wikipedia.org/wiki/Imran_Khan")
soup = BeautifulSoup(htmldata, 'html.parser')
data = ''
corpus=''
for data in soup.find_all("p"):
#     print(data.get_text())
    corpus+=data.get_text()
for data in soup.find_all("table"):
#     print(data.get_text())
    corpus+=data.get_text()
print("Text: ",corpus)
```

# 6. Lemmatize the most frequently occurring 5 adjectives

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
# lemmatizer.lemmatize("scarves")

adjlist=[]
for each in Pos_Tag:
    if(each[1]=='JJR' or each[1]=='JJS' ):
#         print(each)
        adjlist.append(each)
# print(adjlist)

elements_count = {}
# iterating over the elements for frequency
for element in adjlist:
   # checking whether it is in the dict or not
    if element in elements_count:
      # incerementing the count by 1
        elements_count[element] += 1
    else:
      # setting the count to 1
        elements_count[element] = 1
# printing the elements frequencies
# for key, value in elements_count.items():
#    print(f"{key}: {value}")

most_freq_adj=[]
num=0
for w in sorted(elements_count, key=elements_count.get, reverse=True):
    print(w, elements_count[w])
    num+=1
    most_freq_adj.append(w[0])
    if(num==5):
        break
lemmatized_words = [lemmatizer.lemmatize(word) for word in most_freq_adj]
lemmatized_words
```

# 7. Stem the least frequently occurring 5 verbs.

```python
vblist=[]
for each in Pos_Tag:
    if(each[1]=='VB' or each[1]=='VBD' or each[1]=='VBG'or each[1]=='VBN'or each[1]=='VBP' or each[1]=='VBZ'):
#         print(each)
        vblist.append(each)
# print(vblist)

elements_count = {}
# iterating over the elements for frequency
for element in vblist:
   # checking whether it is in the dict or not
    if element in elements_count:
      # incerementing the count by 1
        elements_count[element] += 1
    else:
      # setting the count to 1
        elements_count[element] = 1
# printing the elements frequencies
# for key, value in elements_count.items():
#    print(f"{key}: {value}")

most_freq_vb=[]
num=0
for w in sorted(elements_count, key=elements_count.get):
    print(w, elements_count[w])
    num+=1
    most_freq_vb.append(w[0])
    if(num==5):
        break
        
from nltk.stem.snowball import SnowballStemmer
print(SnowballStemmer.languages)
        
sn_stemmer = SnowballStemmer('english')
stemmed_words= [sn_stemmer.stem(word) for word in most_freq_vb]
stemmed_words
```
